# Subplots
This example focuses on arranging multiple plots in a single figure.  We will start by importing matplotlib and setting a bunch of defaults.

````python
import matplotlib.pyplot as plt
import numpy as np

params = {'font.size'   : 12,
		'legend.fontsize': 12,
		'ytick.major.pad': 8,
}

plt.rcParams.update(params)
````

Next, we will define three data files to read and plot, putting the filenames in a list.

````python
filenames = [ 'rdf_0.txt', 'rdf_1.txt', 'rdf_2.txt']
````

The follow code snippet will loop over the filename list, reading in each, and plotting each on their own separate plot. 

````python
ii = 0  #counter
for filename in filenames:
    r, g_of_r = np.loadtxt(fname=filename, skiprows=1).T
    ax = plt.subplot2grid((len(filenames),1), (ii,0), colspan=1)
    
    ax.plot(r, g_of_r, marker='o',color='Red', markersize=5.0, ls='-', label=filename)
    
    plt.ylabel('g(r)')
    plt.legend(loc='upper right')
    ii = ii+1
    if ii == len(filenames):
        plt.xlabel('r')
    

plt.show()
````

By default, this doesn't look very good. Let's try several ways to adjust. First let's try using the `tight_layout` function. 

````python
ii = 0  #counter
for filename in filenames:
    r, g_of_r = np.loadtxt(fname=filename, skiprows=1).T
    ax = plt.subplot2grid((len(filenames),1), (ii,0), colspan=1)
    
    ax.plot(r, g_of_r, marker='o',color='Red', markersize=5.0, ls='-', label=filename)
    
    plt.ylabel('g(r)')
    plt.xlim(0.5, 3.0)
    plt.ylim(0, 4.0)
    plt.legend(loc='upper right')
    
    ii = ii+1
    if ii == len(filenames):
        plt.xlabel('r')
    
plt.tight_layout()
plt.show()
````

We can explicitly define spacing using `subplots_adjust`. 

```
subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
```
Some suggested values :

```
left  = 0.125  # the left side of the subplots of the figure
right = 0.9    # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.2   # the amount of width reserved for blank space between subplots
hspace = 0.2   # the amount of height reserved for white space between subplots
```

````python
ii = 0  #counter
for filename in filenames:
    r, g_of_r = np.loadtxt(fname=filename, skiprows=1).T
    ax = plt.subplot2grid((len(filenames),1), (ii,0), colspan=1)
    
    ax.plot(r, g_of_r, marker='o',color='Red', markersize=5.0, ls='-', label=filename)
    
    plt.ylabel('g(r)')
    plt.xlim(0.5, 3.0)
    plt.ylim(0, 4.0)
    plt.legend(loc='upper right')
    ii = ii+1
    if ii == len(filenames):
        plt.xlabel('r')
    
plt.subplots_adjust(hspace=0.35)
plt.show()
````

We can remove axis labels for all but the bottom plot.

````python
ii = 0  #counter
for filename in filenames:
    r, g_of_r = np.loadtxt(fname=filename, skiprows=1).T
    ax = plt.subplot2grid((len(filenames),1), (ii,0), colspan=1)
    
    ax.plot(r, g_of_r, marker='o',color='Red', markersize=5.0, ls='-', label=filename)
    
    plt.ylabel('g(r)')
    plt.xlim(0.5, 3.0)
    plt.ylim(0, 4.0)
    plt.legend(loc='upper right')
    
    ii = ii+1
    if ii == len(filenames):
        plt.xlabel('r')
    else:
        ax.set_xticklabels(())
    
plt.subplots_adjust(hspace=0.15)
plt.show()
````

We can also change the size/aspect ratio of the figure itself using `figure(figsize=(x,y))`

````python
ii = 0  #counter


plt.figure(figsize=(4,6))

for filename in filenames:
    r, g_of_r = np.loadtxt(fname=filename, skiprows=1).T
    ax = plt.subplot2grid((len(filenames),1), (ii,0), colspan=1)
    
    ax.plot(r, g_of_r, marker='o',color='Red', markersize=5.0, ls='-', label=filename)
    
    plt.ylabel('g(r)')
    plt.xlim(0.5, 3.0)
    plt.ylim(0, 4.0)
    plt.legend(loc='upper right')
    
    ii = ii+1
    if ii == len(filenames):
        plt.xlabel('r')
    else:
        ax.set_xticklabels(())
plt.subplots_adjust(hspace=0.15)
plt.show()
````

And just as easily change the arrangement as well by changing `subplot2grid`

````python
ii = 0  #counter


plt.figure(figsize=(10,4))

for filename in filenames:
    r, g_of_r = np.loadtxt(fname=filename, skiprows=1).T
    ax = plt.subplot2grid((1, len(filenames)), (0,ii), colspan=1)
    
    ax.plot(r, g_of_r, marker='o',color='Red', markersize=5.0, ls='-', label=filename)
    
    plt.xlim(0.5, 3.0)
    plt.ylim(0, 4.0)
    plt.legend(loc='upper right')
    plt.xlabel('r')


    if ii == 0 :
        plt.ylabel('g(r)')
    else:
        ax.set_yticklabels(())
    ii = ii+1



plt.subplots_adjust(wspace=0.15)
plt.show()
````
